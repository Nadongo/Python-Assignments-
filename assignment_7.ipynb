```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris

#enabling inline plotting for jupyter
%matplotlib inline
```

```python
#Task 1: loading and exploring the dataset
#loading iris data from sklearn
try:
    from sklearn.datasets import load_iris  # Added import statement
    import pandas as pd  # Added import for pandas
    
    iris = load_iris()
    df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
    df['species'] = iris.target_names[iris.target]  # Changed minus (-) to equals (=)
   
except Exception as e:
    print(f"Error loading dataset: {e}")

print("First 5 rows of the dataset are as below:")
df.head()
```

```python
#exploring the dataset more
print("\nDataset Information:")
df.info()

print("\nChecking for any missing values:")
print(df.isnull().sum())
```

```python
#TASK 2: performing basic data analysis
print("\nBasic Statistics:")
print(df.describe())

print("\nMean of each feature per species:")
df.groupby('species').mean()

print(df.columns)
```

```python
#Task 3: DATA VISUALIZATION
#load a line chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['petal length (cm)'][:50], label= 'Petal Length')
plt.title('Petal Length Trend (First 50 entries)')
plt.xlabel('Index')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```

```python
#loading a bar graph

plt.figure(figsize=(7, 5))
df.groupby('species')['petal length (cm)'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Petal Length per Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
#Loading a histogram

plt.figure(figsize=(7, 5))
plt.hist(df['sepal width (cm)'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
#loading a scatter plot
import seaborn as sns

plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species')
plt.title('Sepal Length vs. Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# IN CONCLUSION
# - Setosa species has smaller petals than the other species.
# - There is a positive correlation between sepal length and petal length.
# - Petal length clearly separates different species.
```
